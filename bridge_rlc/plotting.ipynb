```python
import matplotlib.pyplot as plt
import h5py
import numpy as np

# Function to load .mat files using h5py
def load_mat_v73(filename):
    with h5py.File(filename, 'r') as f:
        data = {key: f[key][:] for key in f.keys()}
    return data

# Load voltage and current data from .mat files
voltage_data = load_mat_v73('voltage.mat')
current_data = load_mat_v73('current.mat')

# Extract voltage and current values (adjust key names as needed)
voltage = voltage_data['voltage'][1, :]  # Adjust indexing as needed
current = current_data['current'][1, :]  # Adjust indexing as needed

# Create a new figure
plt.figure()

# Plot the data
plt.plot(voltage, current, linewidth=1.5)

# Add title and labels
plt.title('Line Plot', fontsize=14, fontweight='bold')
plt.xlabel('Voltage (V)', fontsize=12, fontweight='bold')
plt.ylabel('Current (A)', fontsize=12, fontweight='bold')

# Add grid
plt.grid(True)

# Set axis properties
plt.tick_params(axis='both', which='major', labelsize=12)

# Show the plot
plt.show()
```
